```python
import simpy
import random

# Maximo de vehiculos que puede recibir el negocio
MAX_Clientes = 100
# Total de cajas
NUM_CAJAS = 1
# Tiempo que tarda en caja (minutos)
TIEMPO_ATENCION = 6 
# Intervalo de tiempo en que llegan los clientes (minutos)
INTERVALO_LLEGADA = 4
# Tiempo de simulación (minutos)
TIEMPO_SIMULACION = 720


class Tienda(object):
    
    def __init__(self, environment, num_cajas, tiempo_atencion):
        self.env=environment
        self.cajas = simpy.Resource(environment, num_cajas)
        self.tiempo_atencion = tiempo_atencion
        
    def atender_cliente(self, cliente):

        yield self.env.timeout(TIEMPO_ATENCION)
        
        print('Atendiendo cliente {%d%%} en caja => %s ' % (random.randint(10,60), cliente))
        


def llegada_cliente(env, nombre, tienda):

    print('Llega cliente: %s a la hora %.2f.' % (nombre, env.now))
    
   
    with tienda.cajas.request() as caja:
       
        yield caja
       
        print('Pasa Cliente a caja: %s a la hora %.2f.' % (nombre, env.now))
        yield env.process(tienda.atender_cliente(nombre))
        print('Cliente [%s] siendo atendido las %.2f.' % (nombre, env.now))
    
    
def ejecutar_simulacion(env, num_cajas, tiempo_atencion, intervalo):
    tienda=Tienda(env, num_cajas, tiempo_atencion)
    for i in range(6):
        env.process(llegada_cliente(env, 'Cliente-%d'%(i+1),tienda))
    
    while True:
        yield env.timeout(random.randint(intervalo-2, intervalo+2))
        i+=1
        env.process(llegada_cliente(env,'Cliente-%d'%(i+1),tienda))

print('Tienda UPS')
random.seed(4)


env=simpy.Environment()
env.process(ejecutar_simulacion(env, NUM_CAJAS, TIEMPO_ATENCION, INTERVALO_LLEGADA))


env.run(until = TIEMPO_SIMULACION)
```
